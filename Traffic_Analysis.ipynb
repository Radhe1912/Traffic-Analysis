```python
import matplotlib.pyplot as plt
import numpy as np

class Analysis:
    
#     def __init__(self):
#         pass
    
    
    
    def city_area(self):
        
        x=["Gandhinagar","Ahemdabad","Surat","Baroda","Rajkot"]
        y=np.array([325,505,475,455,685])
        label_font={"size":20,"color":"blue","family":"arial"}
        title_font={"size":40,"color":"gold","family":"arial"}

        plt.subplot(2,2,1)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Area(in kmsq)",fontdict=label_font)
        plt.title("Analysis using bar plot",fontdict=title_font)
        plt.bar(x,y)

        plt.subplot(2,2,2)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Area(in kmsq)",fontdict=label_font)
        plt.title("Analysis using Line Diagram",fontdict=title_font)
        plt.plot(x,y,marker="*",ms=20)

        plt.subplot(2,2,3)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Area(in kmsq)",fontdict=label_font)
        plt.title("Analysis using Scatter plot",fontdict=title_font)
        plt.scatter(x,y,label="area")
        plt.legend()

        plt.subplot(2,2,4)
        plt.title("Analysis using bar plot",fontdict=title_font)
        plt.pie(y,labels=x)
        plt.legend()
        
        plt.show()
    
    
    
    def city_population(self):
        
        x=["Gandhinagar","Ahemdabad","Surat","Vadodara","Rajkot"]
        y=np.array([500000,8500000,7000000,5300000,1900000])/100000
        
        label_font={"size":20,"color":"blue","family":"arial"}
        title_font={"size":40,"color":"gold","family":"arial"}

        plt.subplot(2,2,1)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Population in lakhs",fontdict=label_font)
        plt.title("Analysis using bar plot",fontdict=title_font)
        plt.bar(x,y)

        plt.subplot(2,2,2)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Population in lakhs",fontdict=label_font)
        plt.title("Analysis using Line Diagram",fontdict=title_font)
        plt.plot(x,y,marker="*",ms=20)

        plt.subplot(2,2,3)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Population in lakhs",fontdict=label_font)
        plt.title("Analysis using Scatter plot",fontdict=title_font)
        plt.scatter(x,y,label="area")
        plt.legend()

        plt.subplot(2,2,4)
        plt.title("Analysis using pie chart",fontdict=title_font)
        plt.pie(y,labels=x)
        plt.legend()
        
        plt.show()
    
    
    
    def city_vehicles(self):
        
        x=["Gandhinagar","Ahemdabad","Surat","Vadodara","Rajkot"]
        y=np.array([500000*0.9,8500000*0.8,7000000*0.8,5300000*0.8,1900000*0.7])//100000
        
        label_font={"size":20,"color":"blue","family":"arial"}
        title_font={"size":40,"color":"gold","family":"arial"}

        plt.subplot(2,2,1)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Vehicles (in lakhs)",fontdict=label_font)
        plt.title("Analysis using bar plot",fontdict=title_font)
        plt.bar(x,y)

        plt.subplot(2,2,2)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Vehicles in lakhs",fontdict=label_font)
        plt.title("Analysis using Line Diagram",fontdict=title_font)
        plt.plot(x,y,marker="*",ms=20)

        plt.subplot(2,2,3)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Vehicles in lakhs",fontdict=label_font)
        plt.title("Analysis using Scatter plot",fontdict=title_font)
        plt.scatter(x,y,label="area")
        plt.legend()

        plt.subplot(2,2,4)
        plt.title("Analysis using pie chart",fontdict=title_font)
        plt.pie(y,labels=x)
        plt.legend()
        
        plt.show()
    
    
    
    def accident_analysis(self):
        
        x=["Gandhinagar","Ahemdabad","Surat","Vadodara","Rajkot"]
        y=np.array([1000,3000,2000,1500,1800])
        z=np.array([900,2850,1800,1300,1650])
        w=y-z
        
        label_font={"size":20,"color":"blue","family":"arial"}
        title_font={"size":40,"color":"gold","family":"arial"}
        
        plt.subplot(2,2,1)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Total Accident",fontdict=label_font)
        plt.title("Total Accidents",loc="center",fontdict=title_font)
        plt.bar(x,y,color="pink")
        
        plt.subplot(2,2,2)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("No. of People Survived",fontdict=label_font)
        plt.title("Total Accidents",loc="center",fontdict=title_font)
        plt.bar(x,z,color="green")
        
        plt.subplot(2,2,3)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("No. of People couldn't Survived",fontdict=label_font)
        plt.title("Total Accidents",loc="center",fontdict=title_font)
        plt.bar(x,w,color="red")
        
        plt.subplot(2,2,4)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Total Accident",fontdict=label_font)
        plt.title("Total Accidents",loc="center",fontdict=title_font)
        plt.barh(x,y,color="pink")
        
        plt.show()



    def all_analysis_using_linediagram(self):

        City=["Gandhinagar","Ahemdabad","Surat","Vadodara","Rajkot"]

        Area=np.array([325,505,475,455,685])

        Population=np.array([500000,8500000,7000000,5300000,1900000])/100000

        Vehicles=np.array([500000*0.9,8500000*0.8,7000000*0.8,5300000*0.8,1900000*0.7])//100000

        Total_accidents=np.array([1000,3000,2000,1500,1800])

        Survived=np.array([900,2850,1800,1300,1650])

        Not_survived=Total_accidents-Survived


        label_font={"size":20,"color":"blue","family":"arial"}
        title_font={"size":40,"color":"gold","family":"arial"}
        suptitle_font={"color":"purple","family":"arial"}
        
        
        plt.subplot(2,2,1)
        plt.plot(City,Area,marker="*",ms=20)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Area",fontdict=label_font)
        plt.title("Area Analysis",fontdict=title_font)
        
        plt.subplot(2,2,2)
        plt.plot(City,Population,marker="*",ms=20)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Population",fontdict=label_font)
        plt.title("Population Analysis",fontdict=title_font)
        
        plt.subplot(2,2,3)
        plt.plot(City,Vehicles,marker="*",ms=20)
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Vehicles",fontdict=label_font)
        plt.title("Vehicle Analysis",fontdict=title_font)
        
        plt.subplot(2,2,4)
        plt.plot(City,Total_accidents,color="pink",label="Total Accidents",marker="*",ms=20)
        plt.plot(City,Survived,color="green",label="Survived",marker="*",ms=20)
        plt.plot(City,Not_survived,color="red",label="Not Survived",marker="*",ms=20)
        plt.legend(loc="upper right")
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Total Accidents",fontdict=label_font)
        plt.title("Accident Analysis",fontdict=title_font)
                
        plt.suptitle("Traffic Analysis",fontdict=suptitle_font,size=60)
        plt.show()
        
        
    def all_analysis_using_barplot(self):

        City=["Gandhinagar","Ahemdabad","Surat","Vadodara","Rajkot"]

        Area=np.array([325,505,475,455,685])

        Population=np.array([500000,8500000,7000000,5300000,1900000])/100000

        Vehicles=np.array([500000*0.9,8500000*0.8,7000000*0.8,5300000*0.8,1900000*0.7])//100000

        Total_accidents=np.array([1000,3000,2000,1500,1800])

        Survived=np.array([900,2850,1800,1300,1650])

        Not_survived=Total_accidents-Survived


        label_font={"size":20,"color":"blue","family":"arial"}
        title_font={"size":40,"color":"gold","family":"arial"}
        suptitle_font={"color":"purple","family":"arial"}
        
        
        plt.subplot(2,3,1)
        plt.bar(City,Area,color="gold")
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Area (in kmsq)",fontdict=label_font)
        plt.title("Area Analysis",fontdict=title_font)
        
        plt.subplot(2,3,2)
        plt.bar(City,Population,color="purple")
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Population (in lakhs)",fontdict=label_font)
        plt.title("Population Analysis",fontdict=title_font)
        
        plt.subplot(2,3,3)
        plt.bar(City,Vehicles,color="cyan")
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Vehicles",fontdict=label_font)
        plt.title("Vehicle Analysis",fontdict=title_font)
        
        plt.subplot(2,3,4)
        plt.bar(City,Total_accidents,color="pink")
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Total Accidents",fontdict=label_font)
        plt.title("Accident Analysis",fontdict=title_font)
        
        plt.subplot(2,3,5)
        plt.bar(City,Survived,color="green")
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Total Survived",fontdict=label_font)
        plt.title("Survived Analysis",fontdict=title_font)
        
        plt.subplot(2,3,6)
        plt.bar(City,Not_survived,color="red")
        plt.xlabel("Cities",fontdict=label_font)
        plt.ylabel("Total of Not Survived",fontdict=label_font)
        plt.title("Not Survived Analysis",fontdict=title_font)
        
        plt.suptitle("Traffic Analysis",fontdict=suptitle_font,size=60)
        plt.show()
```

```python
plt.figure(figsize=(20,20))
Traffic=Analysis()

user_input=int(input('''Enter 1) For Analysis of Cities Area wise
      2) For Analysis of Cities Population wise
      3) For Analysis of Vehicles of cities
      4) For Analysis of Accident In cities
      5) For Analysis of all Parts \n'''))

if user_input==1:
    Traffic.city_area()

elif user_input==2:
    Traffic.city_population()

elif user_input==3:
    Traffic.city_vehicles()

elif user_input==4:
    Traffic.accident_analysis()

elif user_input==5:
    input_for_all_analysis=int(input('''Enter 1) For Analyse using line diagram
          2) For Analyse using bar plot'''))
    if input_for_all_analysis==1:
        Traffic.all_analysis_using_linediagram()
    elif input_for_all_analysis==2:
        Traffic.all_analysis_using_barplot()
    else:
        print("Please enter 1 or 2 only")

else:
    print("Please enter 1 to 5 only")
```
